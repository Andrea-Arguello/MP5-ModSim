```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def random_iteration_algorithm(w, reps, x_0, s, weights): #Descrito por Barnsley
    x_n = [x_0]
    for i in range(reps):
        """
        Escogemos un w al azar, 
        con probabilidades weights
        y escogemos x_i recursivamente
        """
        x_i = np.random.choice(w, 1, p=weights)[0](x_n[-1])
        x_n.append(x_i)
    puntos = np.array(x_n)
    plt.scatter(puntos[:,0], puntos[:,1], s)
    plt.show()
```

```python
equal_weights = lambda w: [1/w for i in range(w)]
```

```python
random.seed(120)
plt.rcParams["figure.figsize"] = (15,10)
f_1 = lambda coord: [0.5*(coord[0]-coord[1]),0.5*(coord[0]+coord[1])]
f_2 = lambda coord: [-f_1(coord)[1]+1,f_1(coord)[0]]
random_iteration_algorithm([f_1, f_2], 50000, [random.random(),random.random()], 10, equal_weights(2))
```

```python
plt.rcParams["figure.figsize"] = (10,10)
f_1 = lambda coord: [0.5*coord[0]+1,0.5*coord[1]+1]
f_2 = lambda coord: [0.5*coord[0]+50,0.5*coord[1]+1]
f_3 = lambda coord: [0.5*coord[0]+50,0.5*coord[1]+50]
random_iteration_algorithm([f_1, f_2, f_3], 50000, [0,0], 1, equal_weights(3))
```
